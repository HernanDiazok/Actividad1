```python
import sys
import os

sys.path.append(os.path.abspath("../src"))
#importamos tres funciones dentro de la carpeta src
from funciones import calcular_puntos, determinar_mer, mostrar_ranking
#importamos la base de datos
from base_de_datos import rondas



equipos = {}  #Diccionario vacio que se ira actualizando en cada ronda con las estadisticas de cada equipo

for i, ronda in enumerate(rondas, start=1): #en cada iteracion, i sera el numero de la ronda y ronda sera el diccionario de la ronda
    print(f"\nüîπ Ronda {i}:")
    for equipo, stats in ronda.items():        #el for recorre cada equipo y sus estadisticas dentro de la ronda

        if equipo not in equipos:            #si el equipo no esta dentro de los equipos, lo agrega con valor cero
            equipos[equipo] = {'innovacion': 0, 'presentacion': 0, 'errores': 0, 'mer': 0, 'puntos': 0}
        # Suma las estad√≠sticas
        equipos[equipo]['innovacion'] += stats['innovacion']
        equipos[equipo]['presentacion'] += stats['presentacion']
        equipos[equipo]['errores'] += 1 if stats['errores'] else 0
        equipos[equipo]['puntos'] += calcular_puntos(stats)

    # Se determina el Mejor Equipo de la Ronda (MER)
    mer = determinar_mer(ronda) 
    puntos_mer = calcular_puntos(ronda[mer])
    equipos[mer]['mer'] += 1  # Aumenta el contador de MERs
    
    print(f"üèÖ Mejor equipo de la ronda: {mer} ({puntos_mer} puntos esta ronda)")    

    # Se muestra el ranking actual
    mostrar_ranking(equipos)


# Se muestra el ranking final
print("\nüèÜ RANKING FINAL üèÜ")
mostrar_ranking(equipos)   
```
